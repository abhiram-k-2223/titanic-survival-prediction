```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/pclass1.csv')
df2 = pd.read_csv('/content/pclass2.csv')
df3 = pd.read_csv('/content/pclass3.csv')
```

```python
df1.head()
```

```python
df1['Sex'] = df1['Sex'].map({'male': 1, 'female': 0})
```

```python
df2['Sex'] = df2['Sex'].map({'male': 1, 'female': 0})
```

```python
df3['Sex'] = df3['Sex'].map({'male': 1, 'female': 0})
```

```python
df1=df1.iloc[:,1:]
df2=df2.iloc[:,1:]
df3=df3.iloc[:,1:]
```

```python
df1 = df1.fillna(df1.mean())
```

```python
a=df1[['Sex','Age']]
a
```

```python
b = df1.Survived
```

```python
from sklearn.model_selection import train_test_split
a_train,a_test,b_train,b_test = train_test_split(a,b,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
a_train = scaler.fit_transform(a_train)
a_train
```

```python
a_test = scaler.transform(a_test)
a_test
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
```

```python
clf.fit(a_train,b_train)
```

```python
pred=clf.predict(a_test)
```

```python
pred
```

```python
print(clf.predict([[1,28]]))
```

```python
from sklearn.metrics import accuracy_score

score = accuracy_score(b_test,pred)*100
score
```
